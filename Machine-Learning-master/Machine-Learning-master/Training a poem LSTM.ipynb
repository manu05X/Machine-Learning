```python
import numpy as np
```

```python

poem= "Jack and Jill went up the hill to fetch a pail of water "
```

```python
poem =poem.lower().split()
```

```python
poem
```

```python
word_to_int = dict((w,i) for i, w in enumerate(poem))
int_to_word = dict((i,w) for i, w in enumerate(poem))

num_inputs = 300
max_len = 5
dataX =[]
dataY =[]
for i in range(num_inputs):
    start = np.random.randint(len(poem)-2)
    end = np.random.randint(start, min(start+max_len, len(poem)-1))
    sequence_in = poem[start:end+1]
    sequence_out = poem[end +1]
    dataX.append([word_to_int[word] for word in sequence_in])
    dataY.append(word_to_int[sequence_out])
    print(sequence_in, '->', sequence_out)
    
```

```python
dataX
```

```python
dataY
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Embedding, LSTM , Dense, Dropout, Bidirectional
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.models import Sequential
from keras.utils import np_utils
from tensorflow.keras import regularizers

import numpy as np
```

```python
X= pad_sequences(dataX, maxlen=max_len , dtype='float32')
```

```python
X = np.reshape(X, (X.shape[0], max_len, 1))
# normalize
X=X/float(len(poem))
# one hot encoding to the output variable
y=np_utils.to_categorical(dataY)
```

```python
y
```

```python
# define the LSTM model
model = Sequential()
model.add(LSTM(32, input_shape=(X.shape[1], 1)))

model.add(Dense(y.shape[1], activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
X.shape[1], 1
```

```python
model.fit(X, y, epochs=200, batch_size=10, verbose=1)
# summarize performance of the model
scores = model.evaluate(X, y, verbose=0)
print('Model Accuracy: %.2f%%' % (scores[1]*100))
```

```python
for i in range(20):
    
    pattern_index = np.random.randint(len(dataX))
    pattern = dataX[pattern_index]
    
    x = pad_sequences([pattern], maxlen=max_len, dtype='float32')
    x= np.reshape(x,(1, max_len , 1))
    x= x/float(len(poem))
    prediction = model.predict(x, verbose=0)
    index= np.argmax(prediction)
    result = int_to_word[index]
    seq_in = [int_to_word[value] for value in pattern]
    print(seq_in , '->', result)
```
