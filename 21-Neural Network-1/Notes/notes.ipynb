5 input features

Hidden Layer 1 has 10 units

Hidden Layer 2 has 6 units

Output layer has 1 unit

Correct Answer

Solution Description

Parameters for Hidden Layer 1 - (5+1)*10 = 60 Parameters for Hidden Layer 2 - (10+1) * 6 = 66 Parameters for Output Layer - (6 + 1) = 7 Total Parameters - 60 + 66 + 7 = 133
