```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
```

```python
data = pd.read_csv("boston_x_y_train.csv", delimiter=",")
Y=data[' Y']
del data[' Y']
data=preprocessing.scale(data)
data=pd.DataFrame(data)
lc=np.ones(data.shape[0])
lc.shape
data['lc']=lc
points=np.array(data)
```

```python
def step_gradient(points, learning_rate, m,c):
    m_slope = np.zeros(14)
    c_slope = np.zeros(14)
    M = len(points)
    for i in range(M):
        x = points[i]
        y = Y[i]
        for j in range(14):
            m_slope[j] = m_slope[j]+(-2/M)* (y - (m * x).sum() - c.sum())*x[j]
            c_slope[j] = m_slope[j]+(-2/M)* (y - (m * x).sum() - c.sum())
    for j in range(14):
        m[j] = m[j] - learning_rate*m_slope[j]
        c[j] = c[j] - learning_rate*c_slope[j]
    return m,c
```

```python
def gd(points, learning_rate, num_iterations):
    m=np.zeros(14)
    c=np.zeros(14)
    for i in range(num_iterations):
        m,c = step_gradient(points, learning_rate, m,c)
        print(i, " Cost: ", cost(points, m,c))
    return m,c
```

```python
def cost(points, m,c):
    total_cost = 0
    M = len(points)
    for i in range(M):
        x = points[i]
        y = Y[i]
        total_cost += (1/M)*((y - (m*x-c).sum() )**2)
    return total_cost
```

```python
def run():
    learning_rate = 0.07
    num_iterations = 900
    m,c = gd(points, learning_rate, num_iterations)
    return m,c
```

```python
m,c=run()
```
