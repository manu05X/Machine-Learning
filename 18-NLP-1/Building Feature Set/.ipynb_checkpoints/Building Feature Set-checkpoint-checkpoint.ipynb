# Loading And Cleaning Data

we will create 2000 size array named document and fill it with tupple containing [(words,category),(),(),........... 1999]

```python
documents=[]
for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids(category):
        documents.append((movie_reviews.words(fileid),category))
```

# lemmatizing (cleaning)

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
```

```python
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

### Stop word removal

```python
from nltk.corpus import stopwords
```

```python
import string
```

```python
stops=set(stopwords.words('english'))
```

```python
punctuations=list(string.punctuation)
```

```python
# adding punctuations to stop word to remove punctuation along with stop word
stops.update(punctuations)
```

```python
stops,string.punctuation
```

# cleaning function

```python
def clean_reviews(words):
    output_words=[]
    for w in words:
        if not w.lower() in stops:
            pos=pos_tag([w])  #pass w as array as pos_tag accept array
            clean_word=lemmatizer.lemmatize(w,get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
documents=[(clean_reviews(document),category) for document,category in documents]
documents[0]
```

## Building Feature Set

```python
#spliting the data for traning and testing
training_documents=documents[:1500] #1st 1500 to be traning i.e 75%
testing_documents=documents[1500:]#last 1500 to be testing i.e 25%
```

```python
all_words=[]
```

```python
#exmple that a + b  append the array 
a = [1,2]
b = [3,4]
a += b
a
```

```python
#doc is 0 bcz doc is tupple, tupple with 0th entry, 0th value to be 
#word 1st value to be catogery, so we will have all word combine in one 
#place, now it is very straight to find frequency distribution
for doc in training_documents:
    all_words+=doc[0]
```

```python
import nltk
```

```python
#finding frequency distribution
freq=nltk.FreqDist(all_words)
freq.most_common(15)
```

```python
#top 3000 word which is most common
common=freq.most_common(3000)
common
```

```python
features=[i[0] for i in common]
```

```python
features
```

```python
documents[0]
```

### creating dictonary:
        if set of word is in array of words then it is true 
        else 
        if set of word is not in array of words then it is false

```python
def get_feature_dict(words):
    current_feature={}
    words_set=set(words)
    for w in features:
        current_feature[w]=w in words_set #if w comes in words_set it is true otherwise it is false,this way put data in dictionary 
    return current_feature
```

```python
#features contain
features
```

```python
#examle: just we see if it is running for training_documents[0][0]
output=get_feature_dict(training_documents[0][0])
output
```

```python
training_data=[(get_feature_dict(doc),category) for doc,category in training_documents]
```

```python
testing_data=[(get_feature_dict(doc),category) for doc,category in testing_documents]
```

## Classification using NLTK Naive Bayes

* data format is created in :
     * Array of tuples where each tuple has a dictionary which basiclly feature and feature value and catogery that the document or data point is part of
     
     
   format of data for NLTK to work upon is: 
   
       * Array[tuple(Features , target)]

       * Feature is a dictonary where we have value corresponding to each feature 

```python
from nltk import NaiveBayesClassifier
classifier=NaiveBayesClassifier.train(training_data)#this will train and then give the classifier
```

```python
nltk.classify.accuracy(classifier, testing_data)
```

```python
#in nltk classify module has accuracy function that take classifier,testing_data parameter
nltk.classify.accuracy(classifier,testing_data)#this give the accuracy of classifer on testing data that we have

# this give most important feature that it have
classifier.show_most_informative_features(15)
```
