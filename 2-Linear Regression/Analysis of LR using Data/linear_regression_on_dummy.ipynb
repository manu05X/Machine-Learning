```python
import numpy as np
```

```python
data = np.loadtxt("data.csv", delimiter=",") # loadind data into numpy
```

```python
data.shape # to know the shape of data
```

```python
 #data[:, 0] means give all the row and 0'th column,reshape make 1 D array into 2 D
#X has all the rows of 0th Col and Y has all the rows of 1st Col
x = data[:, 0].reshape(-1, 1)
y = data[:, 1] #give all the row and 1'st column
```

```python
print(x.shape)
print(y.shape)
```

```python
#Split the data into tranning and testing, 25% into testing and 75% into tranning
#It's going to split the array into four parts, x into 2 and y into 2 
from sklearn import model_selection
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(x, y)
```

```python
#Now create obj of linear regression, it take input as 2 D array so reshape x array
from sklearn.linear_model import LinearRegression
alg1 = LinearRegression()
alg1.fit(X_train, Y_train)
```

```python
# in y = mx + c ,here coef_ is m i.e slope
alg1.coef_
```

```python
# in y = mx + c ,here intercept_ is c i.e constant 
alg1.intercept_
```

```python
#Plot y = mx + c 
import matplotlib.pyplot as plt
m = alg1.coef_[0]
c = alg1.intercept_

x_line = np.arange(0,10,0.1)
y_line = m * x_line + c
plt.plot(x_line, y_line)
plt.show()
```

```python
#plt.scatter(X_train, Y_train) is use to scatter our tranning data points
#plt.scatter(X_train, Y_train), here X_train(75,1) as it is 2 D array we need to convert it into 1 D
# Y_train(75) is a 1 D array it is ok 
import matplotlib.pyplot as plt
m = alg1.coef_[0]
c = alg1.intercept_

x_line = np.arange(0,10,0.1)
y_line = m * x_line + c
plt.plot(x_line, y_line, "r") # making our line red in color "r"
train_1d = X_train.reshape(75)  # reshape into 1-D
plt.scatter(X_train, Y_train)
plt.show()
```

```python
#change the range of our line as it start from 30 to 70 
import matplotlib.pyplot as plt
m = alg1.coef_[0]
c = alg1.intercept_

x_line = np.arange(30,70,0.1)
y_line = m * x_line + c
plt.plot(x_line, y_line, "r")
train_1d = X_train.reshape(75)
plt.scatter(X_train, Y_train)
plt.show()
```

```python
# for testing data point
import matplotlib.pyplot as plt
m = alg1.coef_[0]
c = alg1.intercept_

x_line = np.arange(30,70,0.1)
y_line = m * x_line + c
plt.plot(x_line, y_line, "r")
train_1d = X_test.reshape(25)
plt.scatter(X_test, Y_test)
plt.show()
```

```python
score_test = alg1.score(X_test, Y_test)
score_traning = alg1.score(X_train,Y_train)
score_test, score_traning
```
