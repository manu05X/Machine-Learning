```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = sns.load_dataset('mpg')
df.head()
```

```python
sns.relplot(x = 'horsepower', y = 'mpg', data = df, kind = 'scatter')
plt.show()
```
